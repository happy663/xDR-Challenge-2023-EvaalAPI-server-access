```python
from typing import Dict, List


def calculate_centroid(coefficients: List[Dict[str, float]]) -> Dict[str, float]:
    direction_error_sum = 0.0
    stride_length_error_sum = 0.0
    count = len(coefficients)

    for coefficient in coefficients:
        direction_error_sum += coefficient["direction_error_coefficient"]
        stride_length_error_sum += coefficient["stride_length_error_coefficient"]

    return {
        "direction_error_coefficient": direction_error_sum / count,
        "stride_length_error_coefficient": stride_length_error_sum / count,
    }


# 与えられた係数のリスト
coefficients = [
    {"direction_error_coefficient": 0.9, "stride_length_error_coefficient": 0.8},
    {"direction_error_coefficient": 0.91, "stride_length_error_coefficient": 0.8},
    {"direction_error_coefficient": 0.92, "stride_length_error_coefficient": 0.8},
    {"direction_error_coefficient": 0.93, "stride_length_error_coefficient": 0.8},
    {
        "direction_error_coefficient": 0.9400000000000001,
        "stride_length_error_coefficient": 0.8,
    },
    {
        "direction_error_coefficient": 0.9500000000000001,
        "stride_length_error_coefficient": 0.8,
    },
    {
        "direction_error_coefficient": 0.9600000000000001,
        "stride_length_error_coefficient": 0.8,
    },
    {
        "direction_error_coefficient": 0.9700000000000001,
        "stride_length_error_coefficient": 0.81,
    },
    {
        "direction_error_coefficient": 0.9800000000000001,
        "stride_length_error_coefficient": 0.8300000000000001,
    },
    {
        "direction_error_coefficient": 0.9900000000000001,
        "stride_length_error_coefficient": 0.8,
    },
    {"direction_error_coefficient": 1.0, "stride_length_error_coefficient": 0.8},
    {
        "direction_error_coefficient": 1.0100000000000002,
        "stride_length_error_coefficient": 0.8,
    },
    {"direction_error_coefficient": 1.02, "stride_length_error_coefficient": 0.8},
    {
        "direction_error_coefficient": 1.0300000000000002,
        "stride_length_error_coefficient": 0.8,
    },
    {"direction_error_coefficient": 1.04, "stride_length_error_coefficient": 0.8},
    {
        "direction_error_coefficient": 1.0500000000000003,
        "stride_length_error_coefficient": 0.8,
    },
    {"direction_error_coefficient": 1.06, "stride_length_error_coefficient": 0.8},
    {
        "direction_error_coefficient": 1.0700000000000003,
        "stride_length_error_coefficient": 0.8,
    },
    {"direction_error_coefficient": 1.08, "stride_length_error_coefficient": 0.8},
    {
        "direction_error_coefficient": 1.0900000000000003,
        "stride_length_error_coefficient": 0.8,
    },
    {"direction_error_coefficient": 1.1, "stride_length_error_coefficient": 0.8},
]

# 重心を計算する
centroid = calculate_centroid(coefficients)
print(centroid)
```
